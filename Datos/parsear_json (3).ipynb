```python
#Importamos librerias necesarias
import pandas as pd
import ast
```

```python
#Leemos el archivo json (cambiar el path según corresponda)
with open('australian_user_reviews.json', encoding='utf-8') as json_file:
    #El archivo se vuelve un solo string
    s = json_file.read()
    #Partimos el string por cada salto de linea y removemos espacios en blanco por delante y detrás
    json_string_list = [line.strip() for line in s.splitlines()]
```

```python
#Convertimos la lista de strings en una lista de diccionarios
json_dict_list = [ast.literal_eval(record) for record in json_string_list]
```

```python
#Definimos el orden de las columnas del dataframe
col_order = ['user_id', 'user_url', 'reviews.funny', 'reviews.posted', 'reviews.last_edited', 'reviews.item_id', 'reviews.helpful', 'reviews.recommend']
```

```python
#Normalizamos el json (lista de diccionarios) y agregamos el prefijo "reviews" para identificar las columnas que vienen de la columna anidada
df_json = pd.json_normalize(json_dict_list, record_path='reviews', meta=['user_id', 'user_url'], record_prefix='reviews.')
#Cambiamos el orden de las columnas
df_json = df_json[col_order]
#Reemplazamos valores vacios con None (tambien se puede reemplazar por numpy.nan)
df_json = df_json.replace('', None)
```
