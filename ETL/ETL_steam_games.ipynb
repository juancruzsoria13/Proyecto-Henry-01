```python
#Importacion de librerias
import pandas as pd
import json as js
```

```python
#Almacenamos la ruta del archivo output_steam_games en una variable
pathgames=r"C:\Users\juanc.DESKTOP-LGMDQP1\OneDrive\Documentos\Proyecto Henry 01\Datos\output_steam_games.json"
```

```python
#Creamos un dataframe con el dataset output_steam_games
df_games= pd.read_json(pathgames, lines=True)
```

```python
#Creamos una copia del dataframe para realizarla las transformacioens sobre ellas
df_games_clean=df_games.copy()
```

```python
#Eliminacion de filas que estan completamente nulas
df_games_clean=df_games_clean.dropna(how="all")
```

```python
df_games_clean.head()
```

```python
#Transformamos el formato de los valores de la columna release_date de string a formato de fecha
df_games_clean["release_date"]= pd.to_datetime(df_games_clean["release_date"], errors="coerce")
```

```python
#Creamos una funcion que solo devuelva el año que se lanzo el juego
def solo_año(valor):
    return valor.year

df_games_clean["release_date"]= df_games_clean["release_date"].apply(solo_año)
```

```python
#Convertimo el año en entero sin decimales
df_games_clean["release_date"]=df_games_clean["release_date"].astype("Int64") 
```

```python
#Nulos por columna
df_games_clean.isnull().sum()
```

```python
#Tomamos el primer valor de la lista y lo dejamos como el genero principal. En caso de que el primero sea "free to play" tomamos el segundo valor de la lista
def conserva_uno(lista):
    if isinstance(lista, list):
        if len(lista) > 1 and lista[0].lower() == "free to play":
            return lista[1]
        elif len(lista) > 0:
            return lista[0]
    return None

df_games_clean["genres"] = df_games_clean["genres"].apply(conserva_uno)

# Verificamos los resultados
print(df_games_clean['genres'].head(10))
```

```python
#Primero convertimos todos los string en la palabra Free y luego reemplazamos Free por "0.00" para finalmente convertirlo en float
def co(valor):
    if isinstance(valor,str):
        return "Free"
    else:
        return valor

df_games_clean["price"]= df_games_clean["price"].apply(co)

df_games_clean["price"]= df_games_clean["price"].replace("Free", "0.00")
df_games_clean["price"]= df_games_clean["price"].astype(float)
df_games_clean
```

```python
#Convertir el id a entero
df_games_clean['id'] = df_games_clean['id'].astype('Int64')
```

```python
df_games_clean["specs"].iloc[2]
```

```python
# Columnas a eliminar
columns_to_drop = ['url', 'reviews_url', 'specs', 'early_access',"tags","publisher"]

#Eliminamos las columnas especificadas anteriormente
df_games_clean= df_games_clean.drop(columns_to_drop, axis=1)
```

```python
#Nuevo orden de columnas
new_order= ["id","app_name","developer","genres","price","release_date"]
df_games_clean=df_games_clean[new_order]
```

```python
def trans(valor):
    if isinstance(valor,str):
        return valor.replace('"',"'")
df_games_clean["developer"]=df_games_clean["developer"].apply(trans)
```
