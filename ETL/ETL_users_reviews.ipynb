```python
import pandas as pd
from pandas import json_normalize
import json
import ast as ast
import re
```

```python
path=r"C:\Users\juanc.DESKTOP-LGMDQP1\OneDrive\Documentos\Proyecto Henry 01\Datos\australian_user_reviews.json"
```

```python
lista=[]

with open(path,encoding="UTF-8  ") as archivo:
    for linea in archivo.readlines():
        lista.append(ast.literal_eval(linea))
        
```

```python
df_reviews=pd.DataFrame(lista)
df_reviews
```

```python
df_reviews=df_reviews.drop(["user_url"],axis=1)
```

```python
df_reviews= df_reviews.explode("reviews")
```

```python
df_reviews
```

```python
df_solo_reviews=json_normalize(df_reviews["reviews"])
```

```python
df_reviews.reset_index(drop=True,inplace=True)
```

```python
df_reviews_final=pd.concat((df_reviews.drop(columns=["reviews"]),df_solo_reviews),axis=1)
```

```python
columns_to_drop=["posted","last_edited","funny"]
df_reviews_final=df_reviews_final.drop(columns_to_drop, axis=1)
```

```python
df_reviews_final
```

```python
def extraer_porcentaje(valor):
    if isinstance(valor, str): 
        coincidencia = re.search(r'\((\d+)%\)', valor)
        if coincidencia:
            return coincidencia.group(1)
    return valor


df_reviews_final['helpful'] = df_reviews_final['helpful'].apply(extraer_porcentaje)
```
