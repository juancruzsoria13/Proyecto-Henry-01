```python
import pandas as pd
from pandas import json_normalize
import ast
import json
```

```python
path= r"C:\Users\juanc.DESKTOP-LGMDQP1\OneDrive\Documentos\Proyecto Henry 01\Datos\australian_users_items.json"
```

```python
lista=[]

with open(path, encoding="UTF-8") as archivo:
    for linea in archivo.readlines():
        lista.append(ast.literal_eval(linea))
```

```python
df_items  = pd.DataFrame(lista)
```

```python
df_items= df_items.explode('items')
```

```python
df_items
```

```python
df_items.reset_index(drop=True, inplace=True)
```

```python
df_solo_items = json_normalize(df_items['items'])
```

```python
df_items_final = pd.concat([df_items.drop(columns=['items']), df_solo_items], axis=1)
```

```python
# Lista de columnas a eliminar
columns_to_drop = ['steam_id','user_url','item_name','playtime_2weeks']

# Eliminar las columnas especificadas
df_items_clean = df_items_final.drop(columns=columns_to_drop, errors='ignore')
```
