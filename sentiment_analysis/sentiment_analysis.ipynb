```python


# Inicializar el analizador de sentimientos de NLTK
sia = SentimentIntensityAnalyzer()


def analyze_sentiment(review):
    if review is None or pd.isna(review):
        return 1  # Valor 1 para reseñas ausentes o nulas
    else:
        # Calcular el puntaje de sentimiento utilizando NLTK
        sentiment_score = sia.polarity_scores(review)['compound']
        if sentiment_score < -0.05:
            return 0  
        elif sentiment_score > 0.05:
            return 2  
        else:
            return 1  

df['sentiment_analysis'] = df['review'].apply(analyze_sentiment)


print(df)
```

```python
df=df.drop(["review"], axis=1)
```
